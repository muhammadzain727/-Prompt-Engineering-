```python
import numpy as np
import pandas as pd
import sklearn.preprocessing as pre_process
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
from sklearn.utils import all_estimators
import sklearn.metrics as sm
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix,precision_score, recall_score, f1_score, roc_auc_score
import joblib
```

```python
df=pd.read_csv("cleaned_data.csv")
df.head()
```

```python
X=df.iloc[:,3:14]
X
```

```python
#scalling and transforming
def trans_scale(X_to_trans,encoded_list,scale_list):
    ordinal_encoding=pre_process.OrdinalEncoder()
    standered_scalling=pre_process.StandardScaler()
    label_encoder=pre_process.LabelEncoder()
    transform_x=make_column_transformer( (ordinal_encoding ,encoded_list )
    ,(standered_scalling,scale_list))
    processed_x=transform_x.fit_transform(X)
    return processed_x
```

```python
encoded_list=["stance"]
scale_list=['height_cm', 'reach_in_cm', 'significant_strikes_landed_per_minute', 'significant_striking_accuracy',
             'significant_strikes_absorbed_per_minute', 'significant_strike_defence', 'average_takedowns_landed_per_15_minutes',
               'takedown_accuracy', 'takedown_defense', 'average_submissions_attempted_per_15_minutes']
X=df.iloc[:,3:14]

processed_x=trans_scale(X,encoded_list,scale_list)
print(processed_x)
```

```python
label_enco=pre_process.LabelEncoder()
Y=df["match_results"]
Y=np.c_[Y]
processed_y=label_enco.fit_transform(Y)
processed_y
```

```python
train_x,test_x,train_y,test_y = train_test_split(processed_x , processed_y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(train_x,train_y)
```

```python
pred_y=model.predict(test_x)
pred_y
```

```python
pred_y_Class=label_enco.inverse_transform(pred_y)
predicted_values=pd.DataFrame(pred_y_Class)
test_y_Class=label_enco.inverse_transform(test_y)
test_results=pd.DataFrame(test_y_Class)
```

```python
merged_frame = pd.concat([predicted_values, test_results], axis=1)
merged_frame.columns = ['Predicted', 'Actual']
merged_frame.to_csv("Results.csv")
```

```python
model_test_acc=accuracy_score(test_y,pred_y)
model_test_f1=f1_score(test_y,pred_y,average='weighted')
model_test_precision=precision_score(test_y,pred_y,average='weighted')
model_test_recall=recall_score(test_y,pred_y,average='weighted')
conf_matrix=confusion_matrix(test_y,pred_y)
```

```python
print(f"Accuracy: {model_test_acc}")
print(f"F1 Score: {model_test_f1}")
print(f"Precision: {model_test_precision}")
print(f"Recall: {model_test_recall}")
print() 
print()
print("Confusion Matrix") 
print(conf_matrix)
print()  
```
